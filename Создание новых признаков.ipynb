# Создание новых признаков

### Подргрузим всякого:

```python
import numpy as np
import pandas as pd

import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns
```

### Поработаем с данными и табличкой:

```python
df = pd.read_csv('Data.txt', sep=",", header = None, engine='python')
```

```python
df = df.rename(columns=df.iloc[0])
df = df.drop(index=[0])
df['Год'] = df['Год'].astype(int)
df['Цена(руб)'] = df['Цена(руб)'].astype(int)
```

```python
df['Пробег(км)'] = df['Пробег(км)'].astype(int)
```

### Создадим новый признак: "Пробег в год"

```python
df['Пробег в год'] = df['Пробег(км)'] / (2024 - df['Год'])
```

### Создадим новый признак: "Сколько лощадиных сил в одном см³ объема двигателя:

```python
df['Мощность двигателя (л.с.)'] = df['Мощность двигателя (л.с.)'].astype(float)
df['Объем двигателя (см³)'] = df['Объем двигателя (см³)'].astype(int)
```

```python
df['Л.с. в см³'] = df['Мощность двигателя (л.с.)'] / df['Объем двигателя (см³)']
```

### Создадим новый признак: "Сколько стоит каждый цилиндр в рамках цены мотоцикла"

```python
df['Цилиндров'] = df['Цилиндров'].astype(float)
```

```python
df['Стоимость цилиндра'] = df['Цена(руб)'] / df['Цилиндров']
```
